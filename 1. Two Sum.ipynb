# **1. Two Sum**

#  **Solution**

```python
def twosum(nums, target):
    num_map = {} # Dictionary to store number and its index
    for i, num in enumerate(nums):
        complement = target - num # Find the required pair
        if complement in num_map: # check if complement exists
            return [num_map[complement], i] # Return the indices
        num_map[num] = i # Store the number and its index
    return [] # Should never reach here as per problem statement

# Example test cases
print(twosum([2,7,11,15], 9)) # Output: [0, 1]
print(twosum([3,2,4], 6))     # Output: [1,2]
print(twosum([3,3], 6))       # Output: [0, 1]
```
